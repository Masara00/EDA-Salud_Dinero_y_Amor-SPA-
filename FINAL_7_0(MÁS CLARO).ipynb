COLUMNAS EN NEGATIVO

En muchos valores 'menos es mejor', por lo que los paso a negativo para que en su normalización 'menos sea peor'

```python
# Poner columnas en valores NEGATIVOS
df_final_junto_negativos = df_final_junto

colum_positivas = ['País', 'Indice Felicidad','Parejas Viviendo Juntas (%)','PIB medio ($)', 'Esperanza de Vida media (años)','Indice Desarrollo','Percepción Corrupción']

for i in df_final_junto_negativos:
    
    if i in colum_positivas:
        pass
    else:
        # print(i)
        df_final_junto_negativos[i] = df_final_junto_negativos[i]*-1

# df_final_junto_negativos.head()
```

```python
# NORMALIZACIÓN ENTRE 0 y 1
df_final_junto_negativos_normalizado = df_final_junto_negativos

for i in df_final_junto_negativos:
    if i == 'País':
        pass
    else:
        column = i
        df_final_junto_negativos_normalizado[column] = MinMaxScaler().fit_transform(np.array(df_final_junto_negativos_normalizado[column]).reshape(-1,1)) 
  
df_final_junto_negativos_normalizado.tail()
```

```python
# Rellenar valores NaN con la media de los otros valores de la columna
df_media_NaN = df_final_junto_negativos_normalizado.fillna(df_final_junto_negativos_normalizado.mean())
```

```python
#  Sacar nueva columna de media

column_names = list(df_media_NaN.columns)
df_media_NaN['Nuevo Indice Felicidad']= df_media_NaN[column_names[:]].mean(axis=1)
df_2_Nuevo_Indice_Felicidad = df_media_NaN[['País', 'Nuevo Indice Felicidad']].copy()

# Normalizar la nueva columna entre 0 y 1
df_media_NaN['Nuevo Indice Felicidad'] = MinMaxScaler().fit_transform(np.array(df_media_NaN['Nuevo Indice Felicidad']).reshape(-1,1))

# Redondear toda la tabla a 2 decimales
df_media_NaN_sin_redondeo = df_media_NaN
df_media_NaN=df_media_NaN.round(decimals=2)
```

```python
# Creando tabla con Rankings Viejo y Nuevo
df_FELICIDAD_2 = df_media_NaN[['País', 'Indice Felicidad', 'Nuevo Indice Felicidad']].copy()
df_FELICIDAD_2['Ranking Viejo'] = df_FELICIDAD_2.index+1

df_felicidades = df_FELICIDAD_2.sort_values('Nuevo Indice Felicidad', ascending=False).reset_index(drop=True)
df_felicidades['Ranking Nuevo'] = df_felicidades.index+1
```
